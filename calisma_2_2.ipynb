```python
# Gerekli kütüphanelerimizi çağırıyoruz
# OpenCv
import cv2
# Numpy
import numpy as np
# Bekleme (sleep) için Zamanla ilgili kütüphane
import time
# Çizim Kütüphanesi
import matplotlib.pylab as plt
%matplotlib inline
```

```python
vcap=cv2.VideoCapture(0)
```

```python
for i in range(5):
    ret,image = vcap.read()
    if ret:        
        plt.axis('off')
        plt.imshow(image)
        plt.show()
        cv2.imwrite("resim"+str(i)+".jpg",image)
    time.sleep(1)
```

```python
vcap.release()
```
